```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Churn_Modelling.csv')
```

```python
data
```

```python
data.columns
```

```python
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
data['Exited'].value_counts()
```

```python
x=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
from imblearn.over_sampling import SMOTENC
```

```python
smt=SMOTENC(categorical_features=[1,2])
```

```python
x_train_sm,y_train_sm=smt.fit_resample(x,y)
```

```python
x_train_sm
```

```python
df=x_train_sm.select_dtypes(['int','float'])
```

```python
df
```

```python
cat_feat=x_train_sm.select_dtypes('object')
```

```python
dummies=pd.get_dummies(cat_feat,drop_first=True)
```

```python
dummies
```

```python
new_df=pd.concat([df,dummies],axis=1)
```

```python
new_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(new_df,y_train_sm,test_size=0.25)
```

```python
## Hyper Parameter Optimization

params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
    
}
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import xgboost
```

```python
classifier=xgboost.XGBClassifier()
```

```python
random_search=RandomizedSearchCV(classifier,param_distributions=params,n_iter=5,scoring='roc_auc',n_jobs=-1,cv=5,verbose=3)
```

```python
new_df.info()
```

```python
classifier=xgboost.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.3,
              enable_categorical=False, gamma=0.4, gpu_id=-1,
              importance_type=None, interaction_constraints='',
              learning_rate=0.3, max_delta_step=0, max_depth=10,
              min_child_weight=5, monotone_constraints='()',
              n_estimators=100, n_jobs=8, num_parallel_tree=1, predictor='auto',
              random_state=0, reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
              subsample=1, tree_method='exact', validate_parameters=1,
              verbosity=None)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(classifier,x_train,y_train,cv=10)
```

```python
xgboost=classifier.fit(x_train,y_train)
predictions=xgboost.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,predictions)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt     

ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax); 

ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['Stayed','Exited']); ax.yaxis.set_ticklabels(['Stayed','Exited']);
```

```python
from pycaret.regression import load_model, predict_model
import streamlit as st
import pandas as pd
import numpy as np

model = load_model('deployment_28042020')

def predict(model, input_df):
    predictions_df = predict_model(estimator=model, data=input_df)
    predictions = predictions_df['Label'][0]
    return predictions

def run():

    from PIL import Image
    

    add_selectbox = st.sidebar.selectbox(
    "How would you like to predict?",
    ("Online", "Batch"))

    st.sidebar.info('This app is created to predict unsubscribers')
    st.sidebar.success('https://www.pycaret.org')
    

    st.title("Unsubscription Prediction App")

    if add_selectbox == 'Online':

        CreditScore = st.number_input('CreditScore', min_value=200, max_value=900, value=200)
        Geography = st.selectbox('Geography', ['France', 'Germany','Spain'])
        Gender = st.selectbox('Gender', ['Female','Male'])
        Age = st.number_input('Age', min_value=1, max_value=100, value=25)
        Tenure = st.number_input('Tenure', min_value=1, max_value=10, value=1)
        Balance = st.number_input('Balance', min_value=0, max_value=260000, value=100000)
        NumOfProducts= st.number_input('NumOfProducts', min_value=0, max_value=4, value=0)
        HasCrCard= st.number_input('HasCrCard', min_value=0, max_value=1, value=0)
        IsActiveMember=st.number_input('ActiveMember', min_value=0, max_value=1, value=0)
        EstimatedSalary=st.number_input('EstimatedSalary', min_value=0, max_value=200000, value=100000)
        
        output=""

        input_dict = {'CreditScore' : CreditScore, 'Geography' : Geography, 'Gender' : Gender, 'Age' : Age, 'Tenure' : Tenure, 'Balance' : Balance,'NumOfProducts':NumOfProducts,'HasCrCard':HasCrCard,'IsActiveMember':IsActiveMember,'EstimatedSalary':EstimatedSalary}
        input_df = pd.DataFrame([input_dict])

        if st.button("Predict"):
            output = predict(model=model, input_df=input_df)
            output = '$' + str(output)

        st.success('The output is {}'.format(output))

    if add_selectbox == 'Batch':

        file_upload = st.file_uploader("Upload csv file for predictions", type=["csv"])

        if file_upload is not None:
            data = pd.read_csv(file_upload)
            predictions = predict_model(estimator=model,data=data)
            st.write(predictions)

if __name__ == '__main__':
    run()
```
